```python
# Задание task_04_02_14.
#
# Выполнил: Снежко Е.А
# Группа: АДЭУ-211
# E-mail: egor.snezhkosnezhko@gmail.com


s = float(input("Пробег за 1-й день (км.) = "))
p = float(input("На сколько увеличивается пробег (%) = "))
p = 1 + p/100
x = 1
sp = s 

total = s
 
while x < 10:
   x += 1
   sp *= p
   print("Пробег за {:}-й день: {:.1f} км.".format(x, sp))
   total += sp
print("Суммарный пробег: {:.1f} км.".format(total))
```
